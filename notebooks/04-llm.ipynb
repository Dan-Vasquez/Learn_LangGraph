```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI

gpt41_llm = ChatOpenAI(model="gpt-4.1-mini", temperature=0.7)
response = gpt41_llm.invoke("Hola")
response.pretty_print()
```

```python
gpt3_llm = ChatOpenAI(model="gpt-3.5-turbo-0125", temperature=0.7)
response = gpt3_llm.invoke("Hola")
response.pretty_print()
```

```python
from langchain_core.messages import AIMessage, SystemMessage, ToolMessage, HumanMessage

msg_1 = SystemMessage(content="Eres un asistente de ayuda")
msg_2 = HumanMessage(content="Me llamo juan")
msg_3 = AIMessage(content="Hola Juan, como estas?")
msg_4 = HumanMessage(content="Como me llamo?")

history = [msg_1, msg_2, msg_3, msg_4]
response = gpt41_llm.invoke(history)
response.pretty_print()
```

```python
from langchain.chat_models import init_chat_model

llm = init_chat_model("openai:gpt-4.1-nano", temperature=0.3)
response = llm.invoke(history)
response.pretty_print()
```
