```python
from langchain_openai import ChatOpenAI

gpt41_llm = ChatOpenAI(model="gpt-4.1-mini", temperature=0.3)
response = gpt41_llm.invoke("Hola soy Daniel y mi telefono es +57 312333270")
response.text
```

```python
from pydantic import BaseModel, Field

class ContactInfo(BaseModel):
    """Informacion de Contacto de una persona"""
    name: str = Field(description="El nombre de una persona")
    email: str = Field(description="El email de una persona")
    phone: str = Field(description="El telefono de una persona")
    tone: int = Field(description="El tono de una persona", ge=0, le=100)
    age: int = Field(description="La edad de una persona")
    sentiment: str = Field(description="El sentimiento de la conversacion")

llm_with_strutured_output = gpt41_llm.with_structured_output(schema=ContactInfo)
messages=[
    ("system", "Eres un asistente de ayuda que extrae informacion de una conversacion"),
    ("user", "Hola soy Daniel y mi telefono es +57 312333270"),
]
response = llm_with_strutured_output.invoke(messages)
response
```
